```python
import re
def clean_name(name):
    intermediate = name.lower().replace(' ', '_')
    intermediate = re.sub(r'[^0-9a-zA-Z]+', '_', intermediate).strip()
    
    if intermediate.endswith('_'):
        intermediate = intermediate[:-1]
    
    if intermediate[0].isdigit():
        intermediate = f"a_{intermediate}"
    
    return intermediate

# mylist = list(df['name-id'])
# mylist.append('1040NR')
# [clean_name(name) for name in mylist]

df['id'] = df.apply(lambda row: clean_name(row['name-id']), axis=1)
```

```python
def url_health_and_status_check(text):
    results = []

    # extract URLs from text
    extractor = URLExtract()
    urls = extractor.find_urls(text)

    # remove dots at the end
    urls = set(url[:-1] if url.endswith(".") else url for url in urls)

    # check that URL returns a valid response
    for url in urls:
        response = requests.get(url)
        results.append(
            {"url": url, "status_code": response.status_code,}
        )

    return results
```

```python
regex = r'^(-?(?:[1-9][0-9]*)?[0-9]{4})-(1[0-2]|0[1-9])-(3[01]|0[1-9]|[12][0-9])T(2[0-3]|[01][0-9]):([0-5][0-9]):([0-5][0-9])(\.[0-9]+)?(Z|[+-](?:2[0-3]|[01][0-9]):[0-5][0-9])?$'

import re
match_iso8601 = re.compile(regex).match
def validate_iso8601(str_val):
    try:
        if match_iso8601( str_val ) is not None:
            if str_val.endswith("Z"):
                str_val = str_val.replace("Z", "+00:00")
            value = datetime.fromisoformat(str_val)
            return True
    except:
        pass
    return False
```
